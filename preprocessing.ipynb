```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('finaldataframe.csv')
data
```

```python
data = data[data['move_list'].str.len() >= 6]
data = data[~data['move_list'].str.contains('*', regex=False)]
data
```

```python
df = pd.DataFrame(columns=['state', 'action'])
i = 0
for _, row in data.iterrows():
    move_list = row['move_list']
#     print(move_list)
    prefix = ''
    idx = 1
    prefix+=move_list[0]
    while move_list[idx].islower() == False:
        prefix+=move_list[idx]
        idx+=1
    next_move = move_list[idx]
    idx+=1
    while idx<len(move_list):
        while (move_list[idx].islower() == False):
            next_move+=move_list[idx]
            idx+=1
            if idx==len(move_list):
                break
        df.loc[i] = [prefix] + [next_move]
        i+=1
        prefix += next_move
        if idx<len(move_list):
            next_move = move_list[idx]
            idx+=1
    
```
